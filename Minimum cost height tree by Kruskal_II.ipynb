```
g = {}
g['A'] = {}
g['B'] = {}
g['C'] = {}
g['D'] = {}
g['E'] = {}
g['F'] = {}
g['G'] = {}
g['A']['B'] = 3
g['A']['C'] = 17
g['A']['D'] = 6
g['B']['A'] = 3
g['B']['D'] = 5
g['B']['G'] = 12
g['C']['A'] = 17
g['C']['E'] = 10
g['C']['F'] = 8
g['D']['A'] = 6
g['D']['B'] = 5
g['D']['E'] = 9
g['E']['C'] = 10
g['E']['D'] = 9
g['E']['F'] = 4
g['E']['G'] = 2
g['F']['C'] = 8
g['F']['E'] = 4
g['F']['G'] = 14
g['G']['B'] = 12
g['G']['E'] = 2
g['G']['F'] = 14

########################################################################

def bfs(graph, start):
    queue = []
    queue += graph[start] 
    visit = [start]

    while queue:
        item = queue.pop(0)
        if not item in visit:
            queue += graph[item] 
            visit.append(item)
    return visit

########################################################################
    
nNodes = len(g)
wLists = list()
for key1, item in g.items():
    for key2, w in item.items():
        wLists.append([w, key1, key2])

tmp_ = []
for i in range(len(wLists)):
  tmp_.append(wLists[i][0])
tmp_ = set(tmp_)
tmp_ = list(tmp_)
tmp_ = sorted(tmp_)

list_ = []
for num in tmp_:
  for wList in wLists:
    if wList[0] == num:
      list_.append(wList)

for i in range(0, len(wLists) // 2):
  if list_[i][0] == list_[i+1][0]:
    del list_[i+1]
    
########################################################################

appendEdges = list_
new_g = {}
check_cycle = []

for appendEdge in appendEdges:
  weight = g[appendEdge[1]][appendEdge[2]]
  keys = list(new_g.keys())
  if appendEdge[1] in keys:
    new_g[appendEdge[1]][appendEdge[2]] = weight   
    _tmp = check_cycle[:]
    check_cycle.append(appendEdge[1]), check_cycle.append(appendEdge[2])
    if set(check_cycle) == set(_tmp):
      del new_g[appendEdge[1]][appendEdge[2]]
  else:
    new_g[appendEdge[1]] = {}
    new_g[appendEdge[1]][appendEdge[2]] = weight
    _tmp = check_cycle[:]
    check_cycle.append(appendEdge[1]), check_cycle.append(appendEdge[2])
    if set(check_cycle) == set(_tmp):
      del new_g[appendEdge[1]][appendEdge[2]]
  if len(set(_tmp)) == len(g.keys()):
    weight = g[appendEdge[1]][appendEdge[2]]
    new_g[appendEdge[1]][appendEdge[2]] = weight
    break
    
print(new_g)

```
